### Before working on null or missing values we need to check and remove the outliers in order to avoid misleading values

## Adult Mortality Outliers removal

```python
#Before outliers removal
sns.set(rc={'figure.figsize':(18,5.27)})
sns.boxplot(le["Adult Mortality"])
le["Adult Mortality"].describe()
```

```python
#before outliers
le.shape
```

```python
#outliers removal
first_quartile = le["Adult Mortality"].quantile(.25)
third_quartile = le["Adult Mortality"].quantile(.75)
IQR = third_quartile - first_quartile
new_boundary = third_quartile + 3*IQR
le.drop(le[le["Adult Mortality"]>new_boundary].index,axis = 0, inplace=True) #Dropping rows with outliers
le.shape
```

As we can observe that rows with outliers have been eleminated. 5 such rows with outliers have been eleminated.

```python
# After outliers removal
sns.set(rc={'figure.figsize':(20,5.27)})
sns.boxplot(le["Adult Mortality"])
le["Adult Mortality"].describe()
```

```python
print(le["Adult Mortality"].isnull().sum())
```

```python
le["Adult Mortality"] = le["Adult Mortality"].interpolate(method = 'linear')
le
```

```python
print(le["Adult Mortality"].isnull().sum())
```

As we can see that the above 2 cells indicate the removal of outliers rows and replacement of the `Null` and `missing` values have been successfully done for `Adult Mortality` column.

#### we apply this process to each column of the dataset 

```python
for col in le:
    if le[col].isnull().sum()> 0:
        #Before outliers removal
        sns.set(rc={'figure.figsize':(20,5.27)})
        sns.boxplot(le[col])
        le[col].describe()
        
        le.shape
        
        #outliers removal
        first_quartile = le[col].quantile(.25)
        third_quartile = le[col].quantile(.75)
        IQR = third_quartile - first_quartile
        new_boundary = third_quartile + 3*IQR
        le.drop(le[le[col]>new_boundary].index,axis = 0, inplace=True) #Dropping rows with outliers
        
        
        le[col] = le[col].interpolate(method = 'linear')

        
```

# creating train test split

```python
# Create train and test data
x = le.drop(columns = ['Life expectancy '])
y = le[['Life expectancy ']]
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
# Convert the data type to float32
X = np.array(x).astype('float32')
y = np.array(y).astype('float32')
```

```python
# split the data into test and train sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Training using `Multiple linear regression model`

```python
# using linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

leModel = LinearRegression(fit_intercept = True)
leModel.fit(x_train, y_train)
```

```python
leModel_accuracy = leModel.score(x_test, y_test)
leModel_accuracy
```

```python
print('Linear Model Coefficient (m): ', leModel.coef_)
print('Linear Model Coefficient (b): ', leModel.intercept_)
```

#### Predicting on test set

```python
# Making prediction
y_predict = leModel.predict(x_test)
```

```python
#Plotting predicted results
plt.plot(y_test, y_predict, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y_predict, y_test) # r square
mse = mean_squared_error(y_predict, y_test) #Mean squared error
rmse = mse**0.5 #Root mean squared error
mae = mean_absolute_error(y_predict, y_test) #Mean absolute error

print("R2 Score: ",r2)
print("MSE: ",mse)
print("RMSE: ", rmse)
print("MAE: ",mae)
```
